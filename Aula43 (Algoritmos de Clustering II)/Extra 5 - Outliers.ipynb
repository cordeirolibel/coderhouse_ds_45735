# Outliers em 1 dimensão

```python
#Import
import matplotlib.pyplot as plt
import numpy as np

#Criando dados para o exemplo
edades = np.array([22,22,23,23,23,23,26,27,27,28,30,30,30,30,31,32,33,34,80])
edad_unique, counts = np.unique(edades, return_counts=True)
```

```python
edad_unique
```

```python
#Número de vezes que cada observação aparece
counts
```

```python
#Verificando os dados
sizes = counts*100
colors = ['blue']*len(edad_unique)
colors[-1] = 'red'

plt.axhline(1, color='k', linestyle='--')
plt.scatter(edad_unique, np.ones(len(edad_unique)), s=sizes, color=colors)
plt.yticks([])
plt.show()
```

Do gráfico anterior podemos interpretar que a maioria das idades estão concentradas, entre os 20 e os 35 anos. Por sua vez, há uma amostra isolada com valor 80.

# Outliers em 2 Dimensões

Agora vamos supor que temos 2 variáveis: idade e renda. Para isso, usaremos uma fórmula para traçar um círculo que delimitará os valores discrepantes. É importante ter em mente que os valores que ultrapassam o valor da "média mais 2 desvios padrão" (a área de o círculo) permanecerá em vermelho.

```python
from math import pi
 
salario_anual_miles = np.array([16,20,15,21,19,17,33,22,31,32,56,30,22,31,30,16,2,22,23])
media = (salario_anual_miles).mean()
std_x = (salario_anual_miles).std()*2
media_y = (edades).mean()
std_y = (edades).std()*2
 
colors = ['blue']*len(salario_anual_miles)
for index, x in enumerate(salario_anual_miles):
    if abs(x-media) > std_x:
        colors[index] = 'red'
        
for index, x in enumerate(edades):
    if abs(x-media_y) > std_y:
        colors[index] = 'red'
 
plt.scatter(edades, salario_anual_miles, s=100, color=colors)
plt.axhline(media, color='k', linestyle='--')
plt.axvline(media_y, color='k', linestyle='--')
 
v=media     #y-position of the center
u=media_y    #x-position of the center
b=std_x     #radius on the y-axis
a=std_y    #radius on the x-axis
 
t = np.linspace(0, 2*pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
 
plt.xlabel('Idade')
plt.ylabel('Salario Anual')
plt.show()
```

**Insight**: Dentro do círculo azul, os valores que estão na média e em vermelho os outliers: 3 valores que ultrapassam o desvio padrão em mais de 2 vezes.

# Outliers em 3 Dimensões

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(7,7))
ax = fig.gca(projection='3d')
 
compras_mes = np.array([1,2,1,20,1,0,3,2,3,0,5,3,2,1,0,1,2,2,2])
media_z = (compras_mes).mean()
std_z = (compras_mes).std()*2
 
for index, x in enumerate(compras_mes):
    if abs(x-media_z) > std_z:
        colors[index] = 'red'
 
ax.scatter(edades, salario_anual_miles, compras_mes, s=20, c=colors)
plt.xlabel('Idade')
plt.ylabel('Salario Anual')
ax.set_zlabel('Compras')
 
plt.show()
```
