```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("../bases/IRIS.csv")
```

```python
#Como siempre realizaremos un peque√±o EDA del dataset
df.head()
```

```python
df.shape
```

```python
#FacetGrid - sepal_length vs sepal_width
sns.FacetGrid(df,hue = 'species',height=4 ).map(plt.scatter,'sepal_length','sepal_width').add_legend();
plt.show()
```

```python
#FacetGrid - sepal_length vs petal_length
sns.FacetGrid(df,hue = 'species' ,height = 4).map(plt.scatter,'sepal_length','petal_length').add_legend();
plt.show()
```

```python
#FacetGrid - sepal_length vs petal_width
sns.FacetGrid(df,hue = 'species',height = 4).map(plt.scatter,'sepal_length','petal_width').add_legend();
plt.show()
```

```python
#FacetGrid - sepal_width vs petal_length
sns.FacetGrid(df,hue = 'species' , height = 4).map(plt.scatter,'sepal_width','petal_length').add_legend();
plt.show()
```

```python
#FacetGrid  - sepal_width vs petal_width
sns.FacetGrid(df,hue = 'species' , height = 4).map(plt.scatter,'sepal_width','petal_width').add_legend();
plt.show()
```
