## Se formos fazer um modelo de regressão é importante verificar a matriz de correlação

```python
#HeatMap
plt.rcParams['figure.figsize'] = (12, 8)

corr = data.corr()
matrix = np.triu(corr)

sns.heatmap(corr, annot = True, cmap = "viridis", mask = matrix)
plt.title('Heatmap for the Dataset', fontsize = 20)
plt.show()
```

```python
# tresbps vs target
plt.rcParams['figure.figsize'] = (8, 6)
sns.boxplot(data,x='target', y='trestbps', palette = 'viridis')
plt.title('tresbps vs target', fontsize = 20)
plt.show()
```

```python
# cholestrol vs target
plt.rcParams['figure.figsize'] = (8, 6)
sns.violinplot(data,x='target', y='chol', palette = 'colorblind')
plt.title('Cholestrol vs Target', fontsize = 20, fontweight = 30)
plt.show()
```

```python
sns.boxenplot(data,x="target", y="age", palette = "Set3")
plt.title("Age vs target", fontsize = 20, fontweight = 30)
plt.show()

#https://seaborn.pydata.org/generated/seaborn.boxenplot.html
```

```python
#Age vs chol
sns.scatterplot(x = "age", y = "trestbps", data = data, hue = "target")
plt.title("Age vs target", fontsize = 20, fontweight = 30)
```
