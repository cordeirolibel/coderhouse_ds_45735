# Requests.get 

```python
import requests
import json
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/bttmly/nba/master/data/teams.json"
```

```python
resposta = requests.get(url)
resposta
```

```python
resposta.text
```

```python
texto = resposta.text
json.loads(texto)
```

```python
jsondata = json.loads(texto)
df = pd.DataFrame.from_dict(jsondata)
df.head()
```
